```python
def compute(init_cost):
    cost = init_cost
    cost = cost - (cost * 0.1)
    it = 1
    while cost >= 1000:
        print("After year ", it, ", the cost is", cost)
        cost = cost - (cost * 0.1)
        it = it+1
        
my_cost = int(input("Enter the cost of the motorbike: "))
compute(my_cost)
```

```python
def operate(operand1, operand2,op):
    if op == 1:
        return operand1 + operand2
    elif op == 2:
        return operand1 - operand2
    elif op == 3:
        return operand1*operand2
    elif op == 4:
        return operand1/operand2
    else:
        return "Invalid operator selected"
    
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
op = int(input("Choose a maths operator to use. \n '1' for addition \n '2' for subtraction \n '3' for multiplication \n '4' for division."))
output = operate(num1,num2,op)
print("The output is: ",output)
```
